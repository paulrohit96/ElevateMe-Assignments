# **EDA**

### **Feature Engineering**

```python
df['TenureRatio'] = df['YearsAtCompany'] / df['TotalWorkingYears'].replace(0, np.nan)
```

```python
df['PerformanceScore'] = df['PerformanceRating'] * df['PercentSalaryHike']
```

```python
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

Converting Categorical target variables into numeric format, for ease of model building and correlation analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot count of attrition
sns.countplot(x='Attrition', data=df)
plt.title('Attrition Count')
plt.show()
```

Most employees are labeled as 'No', meaning they did not leave the company. This is a highly imbalanced dataset, with far fewer Yes (attrition cases) than No. This imbalance is important because it can bias machine learning models toward predicting only 'No'.

```python
# Boxplot: Salary vs Attrition
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df)
plt.title('Monthly Income by Attrition')
plt.show()
```

The thick line in the box represents the median.
Employees who left (Yes) had lower median income compared to those who stayed.
Suggests that lower pay may be a factor in attrition. Those dots above the plot are outliers as people with higher salary also left the organisation.

```python
sns.violinplot(x='Attrition', y='JobSatisfaction', data=df)
plt.title('Job Satisfaction and Attrition')
plt.show()
```

wider violin spread means more job satisfaction. from the plot people with yes for attritioin have less wider violin spread as compared to people who said no

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=[17, 25, 35, 45, 55, 65],
                        labels=['18–25', '26–35', '36–45', '46–55', '56–65'])
```

```python
# AgeGroup vs Attrition
plt.figure(figsize=(8,5))
sns.countplot(x='AgeGroup', hue='Attrition', data=df)
plt.title('Attrition by Age Group')
plt.show()
```

Attrition is highest in the 26-35 age group. Low attrition in older groups (46-55 and 56-65)

```python
# Department vs Attrition
plt.figure(figsize=(8,5))
sns.countplot(x='Department', hue='Attrition', data=df)
plt.title('Attrition by Department')
plt.show()
```

Attrition is higher in Research & Development Depatment and lowest in Human Resource department

```python
# Marital Status vs Attrition
sns.countplot(x='MaritalStatus', hue='Attrition', data=df)
plt.title('Attrition by Marital Status')
plt.show()
```

```python
# Overtime vs Attrition
sns.countplot(x='OverTime', hue='Attrition', data=df)
plt.title('Attrition by Overtime')
plt.show()
```

people who said yes to overtime are more prone to say yes to attrition and vice versa

```python
cols_to_drop = [
    'EmployeeCount', 'EmployeeNumber', 'Over18', 'StandardHours',
    'DailyRate', 'HourlyRate', 'MonthlyRate', 'EducationField',
    'StockOptionLevel'
]

df.drop(columns=cols_to_drop, inplace=True)
```

```python
# Correlation with Attrition
corr = df.corr(numeric_only=True)

# Sort correlation with Attrition
print("Top correlated features with Attrition:\n")
print(corr['Attrition_Flag'].sort_values(ascending=False))

# Correlation heatmap
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()
```

# **Strong Positive Correlations (red areas):**

**MonthlyIncome and JobLevel**: These are strongly positively correlated, which makes sense—higher job levels usually have higher salaries.

**TotalWorkingYears and YearsAtCompany**: Also strongly correlated—people who’ve worked longer tend to have more years at the current company.

**Age and TotalWorkingYears**: Older employees have more working experience.

# **Weak/No Correlations (white/light blue areas):**

DistanceFromHome, WorkLifeBalance, and TrainingTimesLastYear show low or no correlation with most other features.

PerformanceRating has weak or no strong correlations with any variable, which may suggest it’s not tightly linked to other employee attributes.
